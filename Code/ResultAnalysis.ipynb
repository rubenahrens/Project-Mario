```python
import neat
import pickle
import matplotlib.pyplot as plt
```

```python
def plot_stats(statistics, names, filename='resultA.png', title="result"):
    """ Plots the population's average and best fitness. """
    generation = range(len(statistics[0].most_fit_genomes))
    best_fitness = []
    for stat in statistics:
        best_fitness.append([c.fitness for c in stat.most_fit_genomes])
    i = 0
    for best in best_fitness:
        plt.plot(generation, best, label=names[i])
        i += 1

    plt.title(title)
    plt.xlabel("Generations")
    plt.ylabel("Fitness")
    plt.grid()
    plt.legend(loc="best")
    plt.savefig(filename)
    plt.show()
    plt.close()
```

```python
## Experiment A: Change the initial network size
```

```python
statistics = []
statistics.append(pickle.load(open('ExpA/A1Result/stats.pkl', 'rb')))
statistics.append(pickle.load(open('ExpA/A2Result/stats.pkl', 'rb')))
statistics.append(pickle.load(open('ExpA/A3Result/stats.pkl', 'rb')))
names = ['unconnected', 'partial 0.1', 'partial 0.5']
```

```python
plot_stats(statistics, names, filename='resultA.png', title="Different network intializations for NEAT")
```
